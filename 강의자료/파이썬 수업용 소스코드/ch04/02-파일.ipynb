```python
# read() - 파일의 모든 내용을 읽음
file = open('test.txt', 'rt', encoding='utf8')
data = file.read()

print(data)

file.close()
```

```python
# readline() - 줄바꿈 문자까지 (한줄만) 내용 읽음
file = open('test.txt', 'rt', encoding='utf8')
data = file.readline()

print(data)

file.close()
```

```python
# readline() - 내용이 읽히지 않을때까지 (마지막줄까지) 내용 읽음
file = open('test.txt', 'rt', encoding='utf8')
while True:
  data = file.readline()
  if not data:
    break
  print(data, end='')
file.close()
```

```python
# readlines() - 줄바꿈 문자를 기준으로 모든 내용을 리스트 요소로 생성
file = open('test.txt', 'rt', encoding='utf8')
data = file.readlines()
print(data)
file.close()
```

```python
# readlines() - 모든 내용을 리스트로 읽은 후 요소별 출력
file = open('test.txt', 'rt', encoding='utf8')
data = file.readlines()
for line in data:
  print(line, end='')
file.close()
```

```python
# write() - 파일에 내용 쓰기
file = open('write.txt', 'wt', encoding='utf8')

file.write('파일에 저장될 문자\n줄바꿈')

file.close()
```

```python
# writelines() - 파일에 내용 쓰기
file = open('write2.txt', 'wt', encoding='utf8')

file.writelines(['파일에 저장될 문자', '\n', '줄바꿈'])

file.close()
```

### 자동 close

```python
# 일반적인 방법
file = open('song.txt', 'rt', encoding='utf8')
song = file.read()
print(song)
file.close()
```

```python
# with 구문을 사용해서 자동으로 close 하는 방법
file = open('song.txt', 'rt', encoding='utf8')
with file:
  song = file.read()
  print(song)
```

```python
with open('song.txt', 'rt', encoding='utf8') as file:
  song = file.read()
  print(song)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 텍스트 파일 읽기
with open("취임사_제16대.txt", encoding="utf-8") as f:
  text = f.read()

import os
# 후보 폰트 리스트 (운영체제별)
font_candidates = [
    "/System/Library/Fonts/Supplemental/AppleGothic.ttf",  # macOS
    "/usr/share/fonts/truetype/nanum/NanumGothic.ttf",     # Linux
    "C:/Windows/Fonts/malgun.ttf"                         # Windows
]

# 설치된 폰트 중 첫 번째 찾기
font_path = None
for font in font_candidates:
    if os.path.exists(font):
        font_path = font
        break

if font_path is None:
    raise RuntimeError("한글 폰트를 찾을 수 없습니다. 시스템에 폰트를 설치하세요.")

# 워드 클라우드 생성
wc = WordCloud(font_path=font_path, background_color="white", width=800, height=400).generate(text)

# 출력
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
